```python
fig,ax=plt.subplots()
QPC=pixelarrayQPC(distance=5,t=0.5)
QPC.U0=0
QPC.set_all_pixels(0)
sweep=np.linspace(-15,-5,100)
res=[]
for outer_gates in sweep:

    QPC.V1=outer_gates
    QPC.V11=outer_gates
    res.append(QPC.transmission())
    # ax2=QPC.wave_func()
    # ax2.set_title(f'{outer_gates:.2f},{res[-1]:.2f}')
ax.plot(sweep,res)
QPC.wave_func()
```
